```python
import pandas as pd

# Load the dataset
data = pd.read_csv("fifa21_male2.csv")


pd.options.display.max_columns = None
data.head()
```

```python
# Drop irrelevant columns
data = data.drop(["ID", "Player Photo", "Flag Photo", "Club Logo", "Loan Date End"], axis=1)

# Convert preferred foot to a numerical value (0 for left, 1 for right)
data["foot"] = data["foot"].apply(lambda x: 0 if x == "Left" else 1)

# Convert work rate to two numerical values (attack and defense)
#data["Work Rate Attack"] = data["A/W"].apply(lambda x: x.split("/")[0])
#data["Work Rate Defense"] = data["D/W"].apply(lambda x: x.split("/")[1])
#data = data.drop("Work Rate", axis=1)

# Drop rows with missing values
data = data.dropna()
```

```python
import itertools

# Define the possible formations
formations = {
    "4-4-2": [4, 4, 2],
    "4-3-3": [4, 3, 3],
    "3-5-2": [3, 5, 2],
    "3-4-3": [3, 4, 3],
    "5-4-1": [5, 4, 1]
}

# Define the team attributes we want to optimize for (pace, shooting, passing, dribbling, defending, physicality)
attributes = ["PAC", "SHO", "PAS", "DRI", "DEF", "PHY"]

# Define a function to calculate the team rating for a given formation
def calculate_team_rating(formation):
    ratings = []
    for attribute in attributes:
        # Sort players by attribute and choose the top N players for each position
        N = formations[formation][0] if attribute in ["PA", "SHO", "DRI"] else formations[formation][1]
        positions = ["ST", "LW", "RW", "CAM", "CM", "CDM", "LB", "RB", "CB", "GK"]
        top_players = []
        for position in positions:
            position_data = data[data["Position"] == position]
            top_players.extend(position_data.nlargest(N, attribute)["OVA"].tolist())
        # Calculate the average rating for the top players
        rating = sum(top_players) / len(top_players)
        ratings.append(rating)
    # Calculate the overall team rating as the average of the attribute ratings
    overall_rating = sum(ratings) / len(ratings)
    return overall_rating

# Find the best formation
best_formation = None
best_rating = 0
for formation in formations.keys():
    rating = calculate_team_rating(formation)
    if rating > best_rating:
        best_formation = formation
        best_rating = rating

# Print the best formation and rating
print(f"Best formation: {best_formation} (rating: {best_rating:.2f})")
```
